```python
import numpy as np
import random

# ==============================
# CONFIGURACIÓN DEL ENTORNO
# ==============================

FILAS = 4
COLUMNAS = 4
TOTAL_ESTADOS = FILAS * COLUMNAS
TOTAL_ACCIONES = 4  # 0=arriba, 1=abajo, 2=izquierda, 3=derecha

ESTADO_INICIAL = 0
ESTADO_META = 15

# ==============================
# PARÁMETROS DE APRENDIZAJE
# ==============================

TASA_APRENDIZAJE = 0.8
IMPORTANCIA_FUTURO = 0.95

PROB_EXPLORAR = 0.1   # probabilidad de probar algo aleatorio

EPISODIOS = 5000

# ==============================
# TABLA Q (memoria del agente)
# ==============================

tabla_q = np.zeros((TOTAL_ESTADOS, TOTAL_ACCIONES))

# ==============================
# FUNCIÓN DE MOVIMIENTO
# ==============================

def mover(estado, accion):
    fila = estado // COLUMNAS
    columna = estado % COLUMNAS

    if accion == 0:   # arriba
        fila = max(fila - 1, 0)
    elif accion == 1: # abajo
        fila = min(fila + 1, FILAS - 1)
    elif accion == 2: # izquierda
        columna = max(columna - 1, 0)
    elif accion == 3: # derecha
        columna = min(columna + 1, COLUMNAS - 1)

    nuevo_estado = fila * COLUMNAS + columna

    # ==============================
    # RECOMPENSAS Y CASTIGOS
    # ==============================
    if nuevo_estado == ESTADO_META:
        recompensa = 1        # premio grande
    else:
        recompensa = -0.01    # castigo por cada paso

    return nuevo_estado, recompensa

# ==============================
# ENTRENAMIENTO
# ==============================

for episodio in range(EPISODIOS):

    estado_actual = ESTADO_INICIAL

    while estado_actual != ESTADO_META:

        # decidir acción (explorar o usar conocimiento)
        if random.uniform(0, 1) < PROB_EXPLORAR:
            accion = random.randint(0, TOTAL_ACCIONES - 1)
        else:
            accion = np.argmax(tabla_q[estado_actual])

        nuevo_estado, recompensa = mover(estado_actual, accion)

        # actualizar conocimiento (Q-learning)
        mejor_valor_futuro = np.max(tabla_q[nuevo_estado])

        valor_actual = tabla_q[estado_actual, accion]

        nuevo_valor = valor_actual + TASA_APRENDIZAJE * (
            recompensa + IMPORTANCIA_FUTURO * mejor_valor_futuro - valor_actual
        )

        tabla_q[estado_actual, accion] = nuevo_valor

        estado_actual = nuevo_estado

# ==============================
# PRUEBA FINAL (SIN EXPLORAR)
# ==============================

estado = ESTADO_INICIAL
camino = [estado]

while estado != ESTADO_META:
    accion = np.argmax(tabla_q[estado])
    estado, _ = mover(estado, accion)
    camino.append(estado)

print("Camino aprendido:", camino)
```

```python
def obtener_ruta(estado_inicial):
    estado = estado_inicial
    ruta = [estado]

    while estado != ESTADO_META:
        accion = np.argmax(tabla_q[estado])
        estado, _ = mover(estado, accion)
        ruta.append(estado)

    return ruta


# 🔥 PRUEBA
print(obtener_ruta(2))
```
